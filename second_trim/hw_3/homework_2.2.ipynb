```python
import pandas as pd
data = pd.read_csv('form.csv', parse_dates=[0])
```

```python
data.dtypes
```

```python
data.columns = ['time', 'sex', 'child_offence', 'vegan', 'dvach', 'gender', 'rest', 'games', 'series', 'books', 'antisemetism', 'subject', 'crimea', 'putin', 'english', 'old_days', 'films']

data.loc[data['sex'] != 'Девочка', ['sex']] = 0
data.loc[data['sex'] == 'Девочка', ['sex']] = 1

data.loc[data['rest'] != 'Сычевать', ['rest']] = 0
data.loc[data['rest'] == 'Сычевать', ['rest']] = 1
```

```python
child_offence = pd.get_dummies(data.child_offence)
dvach = pd.get_dummies(data.dvach)
gender = pd.get_dummies(data.gender)
antisemetism = pd.get_dummies(data.antisemetism)
subject = pd.get_dummies(data.subject)
crimea = pd.get_dummies(data.crimea)
english = pd.get_dummies(data.english)


child_offence.columns = ['cf1', 'cf2', 'ch3']
dvach.columns = ['dv1', 'dv2', 'dv3', 'dv4']
gender.columns = ['gen1', 'gen2', 'gen3']
antisemetism.columns = ['asemit1', 'asemit2', 'asemit3']
subject.columns = ['subj1', 'subj2', 'subj3', 'subj4', 'subj5', 'subj6', 'subj7', 'subj8']
crimea.columns = ['cr1', 'cr2', 'cr3']
english.columns = ['eng1', 'eng2', 'eng3', 'eng4', 'eng5', 'eng6']


data = data.drop('child_offence', axis=1).join(child_offence)
data = data.drop('dvach', axis=1).join(dvach)
data = data.drop('gender', axis=1).join(gender)
data = data.drop('antisemetism', axis=1).join(antisemetism)
data = data.drop('subject', axis=1).join(subject)
data = data.drop('crimea', axis=1).join(crimea)
data = data.drop('english', axis=1).join(english)
```

```python
data.head()
```

```python
def putin_parse(row):
    if row.putin == 'Политик, лидер и боец':
        return pd.Series([1, 0])
    return pd.Series([0, 1])

data[['putin_good', 'rebel']] = data.apply(putin_parse, axis=1)
data.drop('putin', axis = 1, inplace=True)
```

```python
data.head()
```

```python
# Составляем список всех доступных жанров
ganres = set()
for i, row in data.iterrows():
    for ganre in row.films.split(','):
        ganres.add(ganre.strip())
        

# Создаём колонки под фильмы        
for g in ganres:
    data[g] = 0


# Запоняем колонки
for i, row in data.iterrows():
    for ganre in row.films.split(','):
        data.loc[i, ganre.strip()] = 1

# Переименовываем колонки
data.drop('films', axis=1, inplace=True)
old_names = sorted(list(ganres))
new_names = sorted(['anine', 'triller', 'detective', 'drama', 'historical', 'comedy', 'melodrama', 'tragedy', 'fantasy', 'horror'])

for i, g in enumerate(old_names):
    data.rename(columns={g: new_names[i]}, inplace=True)
```
